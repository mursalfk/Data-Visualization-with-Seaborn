```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
kashti.head()
```

```python
p1=sns.boxplot(x="survived",y="age",showmeans=True,data=kashti)
```

```python
sns.boxplot(x="survived",y="age",showmeans=True,meanprops={"marker":"+","markersize":"12","markeredgecolor":"red"},data=kashti)
plt.xlabel=("How many survived",size=14),
plt.ylabel=("Age {years}",size=14),
plt.title=("Box plot of kitnay dooby or kitnay bach gaye",size=14,weight="bold")
plt.show()
```
